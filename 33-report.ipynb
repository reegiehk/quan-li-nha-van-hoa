## 3. Luồng hoạt động của mô hình hệ thống

### Khởi động
1. Flask app.py khởi động
2. Load seed files db_*.json
3. Copy sang runtime files *.json
4. Load service modules nếu có

### Frontend Client
Người dùng mở browser: http://127.0.0.1:5000/ (mặc định: nha_van_hoa.html)

### Người dùng thao tác
- Thêm hộ khẩu, thành viên
- Xóa hộ, thành viên
- Lập lịch, tài sản
- Tra cứu người

### script.js gửi API
fetch('/api/...', method, body)

### app.py xử lý
1. Parse request
2. Gọi service module nếu có
3. Fallback: đọc/ghi *.json trực tiếp
4. Trả JSON response

### script.js render
Cập nhật UI (render lại danh sách)

## 4. Khó khăn và hạn chế

### Khó khăn
- Nhiều bug tiềm ẩn do không có nhiều kinh nghiệm
- Chưa có test case đủ chất lượng
- Code chưa được tối giản và còn khó đọc, dẫn đến khó trao đổi giữa các thành viên

### Hạn chế hiện tại
| Hạn chế | Tác động | Ưu tiên |
|--------|----------|--------|
| Không có validation input | Dễ lỗi, bất an toàn | Cao |
| Script ghi db_*.json mất seed | Mất dữ liệu gốc, không reset được | Cao |
| Race-condition JSON lock | Dữ liệu hỏng khi concurrent request | Trung |
| Không có test | Khó đảm bảo chất lượng | Trung |
| CSS tối giản | UX không professional | Thấp |
| Không có authentication | Mất bảo mật | Trung |
